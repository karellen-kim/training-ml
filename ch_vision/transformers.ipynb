## 이미지 분류

```python
from transformers import pipeline
from IPython.display import Image
```

```python
imageClassifier = pipeline("image-classification")
```

```python
img = "http://image.dongascience.com/Photo/2016/09/14750507361195.jpg"
display(Image(img))
imageClassifier(img)
```

```python
img = "http://www.coffeehole.co.kr/files/2019/03/12/696681ef971b5a66c984ec4dcb8345b4113058.jpg"
display(Image(img))
imageClassifier(img)
```
